**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_v(img_in):
    H, W = img_in.shape
    out = np.zeros_like(img_in, dtype=float)

    # Maske (Gewichte)
    K = [[-1, 0, 1],
         [-1, 0, 1],
         [-1, 0, 1]]

    # nur Innenbereich, damit 3x3 Fenster komplett drin ist
    for i in range(1, H-1):
        for j in range(1, W-1):

            s = 0.0  # hier sammeln wir die Summe

            # Ã¼ber das 3x3 Fenster laufen
            for a in range(3):          # a = 0,1,2 (Zeile in der Maske)
                for b in range(3):      # b = 0,1,2 (Spalte in der Maske)

                    x = i + (a - 1)     # -1,0,+1 um i herum
                    y = j + (b - 1)     # -1,0,+1 um j herum

                    s = s + img_in[x, y] * K[a][b]  # Pixel * Gewicht

            out[i, j] = s  # Ergebnispixel setzen

    return out

def prewitt_h(img_in):
    H, W = img_in.shape
    out = np.zeros_like(img_in, dtype=float)

    K = [[ 1,  1,  1],
         [ 0,  0,  0],
         [-1, -1, -1]]

    for i in range(1, H-1):
        for j in range(1, W-1):

            s = 0.0

            for a in range(3):
                for b in range(3):
                    x = i + (a - 1)
                    y = j + (b - 1)
                    s = s + img_in[x, y] * K[a][b]

            out[i, j] = s

    return out
```
